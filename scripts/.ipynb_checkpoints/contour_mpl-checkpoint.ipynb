# Contour plot creator

### Preparation
Getting the required libaries

```python
import sys;
import numpy as np
import vtk
import os
from vtk.util.numpy_support import vtk_to_numpy
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import matplotlib.tri as triangulator
#matplotlib.use('PDF') 
```

```python
#Getting the file list from directory
def getListOfFiles(dirName):
    # create a list of file and sub directories 
    # names in the given directory 
    listOfFile = os.listdir(dirName)
    allFiles = list()
    # Iterate over all the entries
    for entry in listOfFile:
        # Create full path
        fullPath = os.path.join(dirName, entry)
        # If entry is a directory then get the list of files in this directory 
        if os.path.isdir(fullPath):
            allFiles = allFiles + getListOfFiles(fullPath)
        else:
            allFiles.append(fullPath)
                
    return allFiles
```

```python
# Reading vtk files
def vtkRead(fileName):
    '''
    Read the data from the file. 
    '''

    reader = vtk.vtkGenericDataObjectReader()
    reader.SetFileName(fileName)
    reader.Update()
    data = reader.GetOutput()
    return (data)
```

```python
#Reading the triangulated vtk scalar data
def triangulateScalarData(data, val):
    '''
    Triangulates the data file and return the grid points in (x, y, z), 
    the connectivity (tri) and the SCALAR values (scalar):
        data:       VTK data output
        x, y, z:    coordinates,
        tri:        triangulated connections,
        val:        scalar value. 
    '''

    cells = data.GetPolys()
    triangles = cells.GetData()
    points = data.GetPoints()
    point_data = data.GetPointData().GetArray(val)

    ntri = int(triangles.GetNumberOfTuples() / 4)
    npts = points.GetNumberOfPoints()
    nvls = point_data.GetNumberOfTuples()

    tri = np.zeros((ntri, 3))
    x = np.zeros(npts)
    y = np.zeros(npts)
    z = np.zeros(npts)
    scalar = np.zeros(nvls)

    for i in range(0, ntri):
        tri[i, 0] = triangles.GetTuple(4 * i + 1)[0]
        tri[i, 1] = triangles.GetTuple(4 * i + 2)[0]
        tri[i, 2] = triangles.GetTuple(4 * i + 3)[0]

    for i in range(points.GetNumberOfPoints()):
        pt = points.GetPoint(i)
        x[i] = pt[0]
        y[i] = pt[1]
        z[i] = pt[2]

    scalar = vtk_to_numpy(data.GetPointData().GetArray(val))
    return (x, y, z, tri, scalar)
```

```python
#Reading triangulated vector data
def triangulateVectorData(data, val):
    '''
    Triangulates the data file and return the grid points in (x, y, z), 
    the connectivity (tri) and the VECTOR values (ux, uy, uz):
        data:       VTK data output, 
        val:        name of vector. 
    Return: 
        x, y, z:    coordinates,
        tri:        triangulated connections,
        ux, uy, uz: vector values. 
    '''

    cells = data.GetPolys()
    triangles = cells.GetData()
    points = data.GetPoints()
    point_data = data.GetPointData().GetArray(val)

    ntri = int(triangles.GetNumberOfTuples() / 4)
    npts = points.GetNumberOfPoints()
    nvls = point_data.GetNumberOfTuples()

    tri = np.zeros((ntri, 3))
    x = np.zeros(npts)
    y = np.zeros(npts)
    z = np.zeros(npts)
    ux = np.zeros(nvls)
    uy = np.zeros(nvls)
    uz = np.zeros(nvls)

    for i in range(0, int(ntri)):
        tri[i, 0] = triangles.GetTuple(4 * i + 1)[0]
        tri[i, 1] = triangles.GetTuple(4 * i + 2)[0]
        tri[i, 2] = triangles.GetTuple(4 * i + 3)[0]

    for i in range(points.GetNumberOfPoints()):
        pt = points.GetPoint(i)
        x[i] = pt[0]
        y[i] = pt[1]
        z[i] = pt[2]

    for i in range(0, point_data.GetNumberOfTuples()):
        U = point_data.GetTuple(i)
        ux[i] = U[0]
        uy[i] = U[1]
        uz[i] = U[2]

    return (x, y, z, tri, ux, uy, uz)
```

```python
#Getting the vector magnitude
def velocityMag(data, nameOfVec):
    '''
    Calculate the magnitude of a vector:
        data:       VTK data output, 
        nameOfVec:  name of the vector field (for instance: 'U').
    Return: 
        x, y, z:    grid coordinates, 
        tri:        connectivity triangles, 
        magnitude:  magnitude of the vector at given grid points. 
    '''

    (x, y, z, tri, u, v, w) = triangulateVectorData(data, nameOfVec)
    magnitude = np.sqrt(u ** 2.0 + v ** 2.0 + w ** 2.0)
    return (x, y, z, tri, magnitude)
```

```python
#Getting the vector magnitude
def uvMag(data, nameOfVec):
    '''
    Calculate the magnitude of a vector:
        data:       VTK data output, 
        nameOfVec:  name of the vector field (for instance: 'U').
    Return: 
        x, y, z:    grid coordinates, 
        tri:        connectivity triangles, 
        magnitude:  magnitude of the vector at given grid points. 
    '''

    (x, y, z, tri, u, v, w) = triangulateVectorData(data, nameOfVec)
    magnitude = np.sqrt(u ** 2.0 + v ** 2.0)
    return (x, y, z, tri, magnitude)
```

```python
#Getting the vector magnitude
def vwMag(data, nameOfVec):
    '''
    Calculate the magnitude of a vector:
        data:       VTK data output, 
        nameOfVec:  name of the vector field (for instance: 'U').
    Return: 
        x, y, z:    grid coordinates, 
        tri:        connectivity triangles, 
        magnitude:  magnitude of the vector at given grid points. 
    '''

    (x, y, z, tri, u, v, w) = triangulateVectorData(data, nameOfVec)
    magnitude = np.sqrt(v ** 2.0 + w ** 2.0)
    return (x, y, z, tri, magnitude)
```

```python
#Getting the vector magnitude
def uwMag(data, nameOfVec):
    '''
    Calculate the magnitude of a vector:
        data:       VTK data output, 
        nameOfVec:  name of the vector field (for instance: 'U').
    Return: 
        x, y, z:    grid coordinates, 
        tri:        connectivity triangles, 
        magnitude:  magnitude of the vector at given grid points. 
    '''

    (x, y, z, tri, u, v, w) = triangulateVectorData(data, nameOfVec)
    magnitude = np.sqrt(u ** 2.0 + w ** 2.0)
    return (x, y, z, tri, magnitude)
```

## Pressure

### Transversal

Getting the list of files and we only need the pressure values and the transversal direction

```python
fileName = getListOfFiles('../postProcessing/triangulatedCuttingPlanes/')
fileName = [el for el in fileName if 'p_T' in el]
fileName
```

```python
#for i in range(0, 1):
for i in range(0, len(fileName)):
    data = vtkRead(fileName[i])
    print(fileName[i], 'loaded')
    
    # select scalars
    x,y,z,tri,scalar = triangulateScalarData(data, "p")
    #x, y, z, tri, ux, uy, uz = d.triangulateVectorData(data, "U")

    # transform data into arrays
    xNew = []
    yNew = []
    zNew = []
    scalarNew = []
    for j in range(0, len(x)):
        yNew.append(y[j] * 1e+3)
        zNew.append(z[j] * 1e+3)
        # scalarNew.append(scalar[j]-348.15)    # scale data for better number formats
        #scalarNew.append(p[j])  # vector component
    x = xNew
    y = yNew
    z = zNew
    #scalar = scalarNew

    # plot parameters
    #xRange = [min(x), max(x)]
    yRange = [min(y), max(y)]
    zRange = [min(z), max(z)]
    xLabel = 'X [mm]'
    yLabel = 'Y [mm]'
    zLabel = 'Z [mm]'
    
    figName=os.path.join('../postProcessing/figures/', '{}{}'.format((os.path.splitext(os.path.basename(fileName[i]))[0]),'.png'))
    

    print(np.min(scalar),np.max(scalar))
    fig1, ax1 = plt.subplots()

    fig_levels = np.linspace(np.min(scalar)-1, np.max(scalar)+1, 20)
    plt.xlim(Tx)
    plt.ylim(Ty)
    plt.rc('font', family='serif', size=5)
    plt.title(os.path.basename(fileName[i]))
    ax1.tricontour(y,z,tri, scalar,fig_levels, colors='black', linewidths=0.1)
    
    ax1.set_aspect('equal')
    ax1.xaxis.set_major_locator(plt.MultipleLocator(50))
    ax1.yaxis.set_major_locator(plt.MultipleLocator(50))
    tcf = ax1.tricontourf(y,z,tri,scalar,fig_levels,cmap=plt.cm.jet, extend='both')
    
    clb=fig1.colorbar(tcf)
    clb.set_label('Static Pressure [Pa]')

    plt.tight_layout()

    #ax1.triplot(y,z,tri, lw=0.2, color='black')
    
    fig1.savefig(figName, format='png', dpi=300)
    
    print('Figure created for', fileName[i])
```
